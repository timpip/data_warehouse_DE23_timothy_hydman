```python
import json
import requests

url = 'https://openparking.stockholm.se/LTF-Tolken/v1/servicedagar/weekday/måndag?outputFormat=json&apiKey=580edc56-114a-4327-8337-733cadaf88c9'
url2 ='https://openparking.stockholm.se/LTF-Tolken/v1/pmotorcykel/all?maxFeatures=100&outputFormat=json&apiKey=580edc56-114a-4327-8337-733cadaf88c9'
       

def _get_ads(url2):
    headers = {'accept': 'application/json'}
    response = requests.get(url2, headers=headers)
    response.raise_for_status()  # check for http errors
    return json.loads(response.content.decode('utf8'))
```
